# **Derivatives**

The derivative of a function f(x) at the point (x, y) is defined as the slope od the tangent line to f(x) at the point (x, y).

SciPy contais the function "[scipy.misc.derivative](https://docs.scipy.org/doc/scipy/reference/generated/scipy.misc.derivative.html#scipy.misc.derivative)", which allows to calculate the derivatives of functions.

Tangent_to_a_curve.png

**Source: Wikipedia**

```python
# calculating the derivative of a function
import numpy as np # import NumPy
from scipy.misc import derivative # import SciPy

def f(x):
  return np.sin(x) # define the function to calculate the derivative

x_point = np.pi/4 # define the point to calculate the derivative

result = derivative(f, x_point, dx=1e-6)

print(f"The derivative of f(x) at x = {x_point} is approximately {result}")
```

# **Integration**

A definite integral of a function can be defined as the signed area of the region bounded by its graph and the horizontal axis.

SciPy allows to calculate integrals using the function "[scipy.integrate.quad](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.quad.html#scipy.integrate.quad)".

Integral_example.png

**Source: Wikipedia**

```python
# calculating the integral of a function
from scipy.integrate import quad

# define the function you want to integrate
def f(x):
    return np.sin(x)

# customize the integration limits
a = 0  # lower limit of integration
b = np.pi  # upper limit of integration

result_integral, error = quad(f, a, b) # returns a tuple: (value of the integral, estimate of the absolute error)

print(f"The integral of f(x) from {a} to {b} is approximately {result_integral} with an error of {error}")
```

For double and triple integrals, you can use scipy.integrate.dblquad and scipy.integrate.tplquad, respectively. These functions allow you to perform nested integrations over multiple variables.

# **Optimisation**

Optimization is the process of finding the best solution from all feasible solutions. In mathematical terms, it involves finding the minimum or maximum value of a function subject to certain constraints. Optimization is widely used in various fields such as economics, engineering, machine learning, and operations research.

SciPy provides various optimization algorithms through the scipy.optimize module. Let's use the "[scipy.optimize.minimize](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html#scipy.optimize.minimize)" function to find the minimum of a simple function.

```python
import numpy as np
from scipy.optimize import minimize

# define the function to be minimized
def f(x):
    return (x - 3)**2

# initial guess for the minimum
x0 = 0  # starting point, can be any number

result = minimize(f, x0, method='BFGS', options={'disp': True})
          # x0: initial guess
          # method: optimization method (e.g., 'BFGS', 'Nelder-Mead', etc.)
          # options: print convergence message

# the result of the optimization contains the optimized value and other information
optimized_x = result.x  # the x value that minimizes the function

function_value_at_minimum = result.fun  # the function value at the minimum

print(f"The minimum value of the function is {function_value_at_minimum} at x = {optimized_x}")
```

Other important feature is the ability of finding the roots of a scalar function or a system of nonlinear equations.

This can be performed with the function "scipy.optimize.root".

For more information, go to: https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.root.html#scipy.optimize.root

```python
from scipy.optimize import root
import numpy as np

# define the function whose root we want to find
def f(x):
    return x**2 - 4

# initial guesses for the roots
initial_guesses = [1, -1]  # starting points close to the expected roots

# list to store the results
roots = []

# loop over initial guesses to find different roots
for x0 in initial_guesses:
    result = root(f, x0, method='hybr')
        # x0: initial guess
        # method: the algorithm to use (e.g., 'hybr', 'lm', 'broyden1', etc.)
    if result.success:
        roots.append(result.x[0])
    else:
        print(f"Root finding did not converge for initial guess {x0}. Message: {result.message}")

# print all found roots
print(f"The roots of the equation are approximately {roots}")
```

**Singular Value Decomposition (SVD)**

In linear algebra, the  **SVD** is a factorization of a real or complex matrix into a rotation, followed by a rescaling followed by another rotation. It generalizes the eigendecomposition of a square normal matrix with an orthonormal eigenbasis to any **m x n** matrix.

Specifically, the singular value decomposition of an **m x n** complex matrix M is a factorization of the form **M = UD$V^{T}$**.

In this form, **U** is an  **m x n**  complex unitary matrix, **D** is an **m x n** rectangular diagonal matrix with the square root of positive einevalues, **V** is an **m x n** complex unitary matrix, and **$V^{T}$** is the conjugate transpose of **V**.

To know more about SVD, go to: https://www.cs.cmu.edu/~venkatg/teaching/CStheory-infoage/book-chapter-4.pdf

Let's use our first matrix to perform SVD.

```python
# SVD function
linalg.svd(matrix)
```

Eigenvalues and Eigenvectors

Eigenvalues and eigenvectors are concepts in linear algebra that are associated with square matrices.

*   **Eigenvectors**: Eigenvectors are non-zero vectors that are mapped to a scaled version of themselves by the linear transformation represented by the matrix.
*   **Eigenvalues**: An eigenvalue of a square matrix **A** is a scalar **b**  such that when the matrix **A** is multiplied by a corresponding eigenvector **v** the result is a scaled version of **v**, i.e., Av = bv.

In SciPy, you can use the "scipy.linalg.eig" function to compute eigenvalues and eigenvectors of a square matrix.

Let's use our first matrix to calculate the Eigevectors and the Eigenvalue.

```python
# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = linalg.eig(matrix)

print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```
