# **Creating Arrays**



In computer science, an array is a data structure consisting of a collection of elements (values or variables), of same memory size, each identified by at least one array index or key. An array is stored such that the position of each element can be computed from its index tuple by a mathematical formula. The simplest type of data structure is a linear array, also called one-dimensional array.

# **Accessing Arrays**



It is possible to access the elements of an array in different ways. You can access an array element (or entire row) by referring to its index number. The indexes in NumPy arrays start with 0, meaning that the first element has an index of 0, the second has an index of 1, etc.

Let us create a 2D array containing the previous arrays partial_distances_1 and partial_distances_2 as distances.

```python
distances = np.array([[15, 25, 35, 45, 88, 77],[25, 98, 45, 77, 36, 18]])
```

This 2D array can be "graphically" represented as:

image.png

So, to access the entire column, we need index [i] and to access a specific element, we need a combination of indexes [i][j]. For example:

```python
# accessing the entire row - first row
distances[0]
```

```python
# accessing the element 36 (second row)
distances[1][4]
```

We can also select all the elements of a column.

```python
# selecting the first column
distances[:,0]
```

We can also access a specifi ranges of values within an array or slicing an array. To do this we need to define where to start and where to stop. For example, we want to access from j = 1 to j = 3.

```python
# slicing an array
distances[:,1:4] # 1 is inclusive but 4 is not
```

We can also iterate through the entire array using a "for loop". For example:

```python
# iterating through an array
for i in distances:
  print(i)
```

Finally, we can also indexing an array with Boolean Arrays. Let us soppose that the limit of distance is 50 and we want to know the elements over 50 of our array distances.

```python
# defines the limit distance
limit_distance = distances>50
```

```python
# find in the array distances those values over the limit
limit_distance
```

```python
# passing limit_distance as an argument to distances
distances[limit_distance]
```

# **Copy and View**

[Copy and Views](https://numpy.org/doc/stable/user/basics.copies.html#) can be interesting resources sometimes.

*   A copy **OWNS** the data, so that changes made to the copy **will not affect** original array, and changes made to the original array **will not affect** the copy.
*   A view **DOES NOT OWN** the data, so that changes made to the view **will affect** the original array, and changes made to the original array **will affect** the view.

Summarising, the copy is a new array, and the view is just a view of the original array.

**View**

```python
# taking our array distances
distances
```

```python
# creating a view of array distances
view_distances = distances.view()
```

```python
# modifing the element 77 to 22
view_distances[1][3]=22
```

```python
# taking the original array
distances
```

**Copy**

```python
# creating a copy of array distances
copy_distances = distances.copy()
```

```python
# modifing the element 22 to 77
copy_distances[1][3]=77
```

```python
# taking the original array
distances
```

```python
# taking the copy
copy_distances
```

# **Shape Manipulation**

It is possible to manipulate the shape of an array by using some functions embedded into NumPy. The main functions to do this are:

*   **[Flatten](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.flatten.html)** - Return a copy of the array collapsed into one dimension ().
*   **[Reshape](https://numpy.org/doc/stable/reference/generated/numpy.reshape.html)** - Gives a new shape to an array without changing its data.
*   **[Resize](https://numpy.org/doc/stable/reference/generated/numpy.resize.html)** - Return a new array with the specified shape.*   
*   **[Stack](https://numpy.org/doc/stable/reference/generated/numpy.stack.html)** - Join a sequence of arrays along a new axis.
*   **[Split](https://numpy.org/doc/stable/reference/generated/numpy.split.html)** - Split an array into multiple sub-arrays as views into ary.

So, let us apply these functions:

```python
# taking our array distances
distances
```

```python
# flatten
flatten_distances = distances.flatten()
flatten_distances
```

```python
# reshape
reshape_distances = distances.reshape(3,4)
reshape_distances
```

```python
# resize
resize_distances = np.resize(reshape_distances,(2,6))
resize_distances
```

```python
# resize 2
resize_distances = np.resize(reshape_distances,(3,7))
resize_distances

# If the new array is larger than the original array, then the new array is filled with repeated copies of the original
```

```python
# split
split_distances = np.split(distances,3)
split_distances
```

```python
# stack
stack_distances = np.stack((np_partial_distances_1,np_partial_distances_2))
stack_distances
```

# **Linear Algebra**

We can also use NumPy to perform [Linear Algebra](https://numpy.org/doc/stable/reference/routines.linalg.html#module-numpy.linalg) operations, such as sum, multiplication, transpose a matrix, calculate the inverse matrix (if exists), calculate the determinant and the trace.

Let us now create a new array, called matrix.

```python
# sum of two marices
matrix1 = np.array[[1, 2],[3, 4]] # define matrix 1
matrix2 = np.array[[5, 6], [7, 8]] # define matrix 2
sum_matrices = matrix1 + matrix2
sum_matrices
```

```python
# multiplying two matrices
A = np.array[[1, 2, 3], [4, 5, 6]] # defines matrix A - 2 rows and 3 columns
B = np.array[[7, 8], [9, 10], [11, 12]] # defines matrix B - 3 rows and 2 columns
C = A.dot(B) # multiply matrix A and B, creating matrix C with 2 rows and 2 columns
C
```

```python
# creating a new array
matrix = np.array([[1.2, 2.6, 3.8],[3.1, 5.6, 7.4],[9.2, 5.4, 1.9]])
matrix
```

```python
# transposing the matrix
transpose_matrix = matrix.transpose()
transpose_matrix
```

```python
# calculating the determinant
determinant_matrix = np.linalg.det(matrix)
determinant_matrix = np.round(determinant_matrix,2)
determinant_matrix
```

```python
# calculating the inverse matrix
inverse_matrix = np.linalg.inv(matrix)
inverse_matrix # only applies to square matrix where determinant is different of zero
```

```python
# calculating the trace (sum of diagonal elements)
trace_matrix = np.trace(matrix)
trace_matrix
```
