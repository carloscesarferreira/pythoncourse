Analyse the Dataset provided by the Fire Department of the New York City (FDNY) to determine:

1. The total number of FDNY facilities.
2. The number of fire service facilities in each borough.
3. List the names of the facilities in Manhattan.

```python
# importing the Pandas library
import pandas as pd
```

```python
# reading the dataset file
path = "https://raw.githubusercontent.com/carloscesarferreira/pythoncourse/main/Season%20II%20%E2%80%93%20Data%20Science%20with%20Python/datasets/FDNY_Firehouse_Listing.csv"
df_fdny_dataset = pd.read_csv(path)
```
